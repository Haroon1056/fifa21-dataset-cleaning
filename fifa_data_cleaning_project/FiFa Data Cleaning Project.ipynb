## 11. Contract
Player's contract details

```python
fifa["Contract"].dtype
```

```python
fifa["Contract"].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row["Contract"] or 'Free' in row["Contract"]:
        print(row["Contract"])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_data = np.nan
        end_data = np.nan
        contract_length = 0
    else:
        start_data, end_data = contract.split(' ~ ')
        start_year = int(start_data[:4])
        end_year = int(end_data[:4])
        contract_length = end_year - start_year
    return start_data, end_data, contract_length


new_cols = ["Contract Start", "Contract End", "Contract Length(years)"]
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa.sample(3)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
# Contract categories
def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'
# add column
fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
```

## 13. Height
Player's height in centimeters

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        # Split the string at the "' " (space after the apostrophe)
        feet, inches = height.split("'")
        total_inches = int(feet) * 12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

# Apply the function to the 'Height' column
fifa['Height'] = fifa['Height'].apply(convert_height)
```

```python
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns={"Height": "Height(cm)"})
```

```python
fifa.sample(3)
```

## 13. Weight
Player's weight in kilograms

```python
fifa["Weight"].dtype
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        # Use a more accurate conversion factor and round to 2 decimal places
        return round(pounds / 2.20462, 2)

# Apply the correct function to the 'Weight' column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
```

```python
fifa["Weight"].unique()
```

```python
fifa = fifa.rename(columns = {"Weight": "Weight(kg)"})
fifa.sample(3)
```

## 20. Value

```python
fifa['Value'].dtype
```

```python
def convert_money(value):
    if isinstance(value, str):
        value = value.replace('€', '').strip()

        if 'M' in value:
            return float(value.replace('M', '')) * 1000000
        elif 'K' in value:
            return float(value.replace('K', '')) * 1000

    return float(value)  # In case it's already a number or needs no conversion
```

```python
fifa['Value'] = fifa['Value'].apply(convert_money)
```

## 77. Hits
Number of times the player has been search in the FiFa database.

```python
fifa['Hits'].dtype
```

```python
def convert_to_int(value):
    # Check if the value is a string and contains 'K'
    if isinstance(value, str) and 'K' in value:
        # Remove the 'K' and multiply by 1000
        return int(float(value.replace('K', '')) * 1000)
    else:
        # Convert the value directly to an integer if it's a string or leave it unchanged if already numeric
        return int(value) if isinstance(value, str) else value

# Apply the function to the relevant column
fifa['Hits'] = fifa['Hits'].apply(convert_to_int)
```
